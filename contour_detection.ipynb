```python
import cv2
import os
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
GX = [[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]]
GY = [[-1, -2, -1], [0, 0, 0], [1, 2, 1]]
THRESHOLD = 250
```

```python
img_input = cv2.imread("input.png")
img_input = cv2.cvtColor(img_input, cv2.COLOR_BGR2GRAY)
plt.imshow(img_input, cmap='gray')
```

```python
(x_size, y_size) = img_input.shape
```

```python
img_output = np.zeros((x_size-2, y_size-2), dtype=np.uint8)
```

```python
for x in range(0, x_size-2):
    for y in range(0, y_size-2):
        gxVal = img_input[x][y] * GX[0][0] + img_input[x][y+1] * GX[0][1] + img_input[x][y+2] * GX[0][2] + img_input[x+1][y] * GX[1][0] + img_input[x+1][y+1] * GX[1][1] + img_input[x+1][y+2] * GX[1][2] + img_input[x+2][y] * GX[2][0] + img_input[x+2][y+1] * GX[2][1] + img_input[x+2][y+2] * GX[2][2]
        gyVal = img_input[x][y] * GY[0][0] + img_input[x][y+1] * GY[0][1] + img_input[x][y+2] * GY[0][2] + img_input[x+1][y] * GY[1][0] + img_input[x+1][y+1] * GY[1][1] + img_input[x+1][y+2] * GY[1][2] + img_input[x+2][y] * GY[2][0] + img_input[x+2][y+1] * GY[2][1] + img_input[x+2][y+2] * GY[2][2]
        img_output[x][y] = 0 if math.sqrt(gxVal ** 2 + gyVal ** 2) < THRESHOLD else 255
```

```python
plt.imshow(img_output, cmap='gray')
```
